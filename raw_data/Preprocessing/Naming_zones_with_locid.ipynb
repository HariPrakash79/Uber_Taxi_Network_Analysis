```python
import pandas as pd
import os

# === CONFIG ===
zone_file = r"D:\MSDS\Network Science\project\nyc-taxi-data-master\analysis\2017_update\data\taxi_zones_simple.csv"
input_folder = r"D:\MSDS\Network Science\project\yellowtrips_cleaned_per_year"
output_folder = r"D:\MSDS\Network Science\project\yellowtrips_with_zones"
os.makedirs(output_folder, exist_ok=True)

# Load zone lookup
zone_lookup = pd.read_csv(zone_file)

# Loop through years
for year in range(2015, 2025):
    file_path = os.path.join(input_folder, f"yellow_tripdata_{year}_cleaned.csv")
    if os.path.exists(file_path):
        print(f"Processing {file_path}")
        df = pd.read_csv(file_path)

        # Merge PU
        df = df.merge(
            zone_lookup[['locationid', 'zone']].rename(columns={'locationid': 'PULocationID', 'zone': 'PU_Zone'}),
            on='PULocationID',
            how='left'
        )

        # Merge DO
        df = df.merge(
            zone_lookup[['locationid', 'zone']].rename(columns={'locationid': 'DOLocationID', 'zone': 'DO_Zone'}),
            on='DOLocationID',
            how='left'
        )

        # Save
        out_file = os.path.join(output_folder, f"yellow_tripdata_{year}_cleaned_with_zones.csv")
        df.to_csv(out_file, index=False)
        print(f"✅ Saved {out_file}")
    else:
        print(f"❗ File not found: {file_path}")
```
